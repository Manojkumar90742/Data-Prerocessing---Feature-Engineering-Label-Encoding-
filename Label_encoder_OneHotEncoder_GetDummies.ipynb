***ENCODING***


*why we do encoding process?*


*we must be translate the string values into integer value.then only the machine will be understand the pattern of the dataset.we have two different types of encoding techniques.one is label encoder and another is one hot encoder technique.*

***LABEL ENCODING***


*In the dataset the given feature values are two unique categorial value or ordinal values we do label encoding technique.*


***Eg:***


          *Two unique categorial values like:True\false,Male\Female,Yes\No

          *Ordinal values like:monday/tuesday/wednessday/thuresday/friday/... or season+ or stage1,stage2,stage3...

*We have two different way to process the label encoding technique.one is by using pandas and another one is import sklearn library to process the technique.*          


```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df['sex']=le.fit_transform(df['sex'])
df['smoker']=le.fit_transform(df['smoker'])
```

```python
df            # 0 means absence of something and 1 means presents of something.
              # Here in sex column 0 means female and 1 means male.
              # And in smoker feature 1 means Yes and 0 means No.
```

***ONE HOT ENCODER***


*In the dataset the given feature values are more then two unique values or nominal value we do one hot encoder technique*


***Eg:***
           
          *More then two unique features like:Male,Female,Transgender.
          *nominal values like:Milk,Water,Oil,Ghee.

*We have two different way to process the one hot encoding technique.one is by using pandas and another one is import sklearn library to process the technique.*          

```python
# import onehotencoder from sklearn library
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder()
```

```python
df_ohe=pd.DataFrame(ohe.fit_transform(df[['region']]).toarray())
df_ohe                                                           # southwest=3,southeast=2,northwest=1,northeast=0
```

```python
df=pd.concat([df,df_ohe],axis=1)
```

```python
df=df.drop(['region'],axis=1)
```
